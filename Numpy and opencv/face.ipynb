```python
from PIL import Image
import matplotlib.pyplot as plt
import cv2
```

```python
frame = cv2.imread('Chris_Hemsworth_as_Thor.jpg') # opencv reads the image as GBR format which is opposite of RGB hence we need to convert it into RGB format
plt.imshow(frame)
```

```python
rgb = cv2.cvtColor(frame,cv2.COLOR_BGR2RGB) # converting the GBR image to RGB
plt.imshow(rgb)
```

```python
classifier = cv2.CascadeClassifier("haarcascade_frontalface_default.xml") #harrcascade is the algorithm for determinng the face components and CascadeClassifier is the function which takes a classification and on the basis of that we can get the different properties of the image.
classifier
```

```python
faces = classifier.detectMultiScale(frame) # through detectScale we are getting the points from that image read and it returns X, Y , WIDTH , HEIGHT co-ordinates.
```

```python
face = faces[0] # if there are multiple faces then we are taking a particular image from that
x,y,w,h = face
```

```python
face = rgb[y:y+h,x:x+w] # from the whole readings we are taking only the face co-ordinates 
plt.imshow(face)
```

```python
out = cv2.rectangle(rgb,(x,y),(x+w,y+h),(0,0,255),1)
plt.imshow(out)

```
