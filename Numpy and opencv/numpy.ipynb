```python
from PIL import Image
import matplotlib.pyplot as plt
```

```python
pil = Image.open("RGB_Matte.png")
```

```python
arr = np.array(pil)
```

```python
arr.shape
```

```python
red = arr[:,:,1] # thid dimension means b/w red,green and blue 0 -> red, 1 -> green and 2 -> blue
# in 1-d list we slice the list like [start:end] so it is being followed for every dimension here we are having 3 dimensions hence we are doing it 3 time and putting , after each slice.
plt.imshow(red)
```

```python
thorimage = Image.open('Chris_Hemsworth_as_Thor.jpg')
thorArr = np.array(thorimage)
plt.imshow(thorArr)
val = thorArr.mean(axis=2) #  y axis -> axis=0, x axis -> axis=1 and z axis ->  axis=2 and z-axis contains color hence we are taking a mean of them
plt.imshow(val,cmap="gray") # if we don't use cmpa then also it will use grayscale but multicolor grayscale we are using cmap which color map though this we are mapping the color with the color proivided
```
